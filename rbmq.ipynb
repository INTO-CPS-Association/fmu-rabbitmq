```python
import pandas as pd
import plotly.graph_objs as go
import glob

files = glob.glob("log/*.csv")
cols = ['simtime', 'stepdur']

obs = 0
l = []
for f in files:
    l.append(pd.read_csv(f, usecols=cols, sep=',' ))
    obs = obs + 1

df = pd.concat(l)
dfm = df.groupby(level=0).mean()
dfmax = df.groupby(level=0).max()
dfmin = df.groupby(level=0).min()

fig = go.Figure([
#     go.Scatter(
#         x=df['simtime'],
#         y=df['stepdur'],
#         mode='markers',
#         name='stepdur'
#     ),
    go.Scatter(
        x=dfm['simtime'],
        y=dfm['stepdur'],
        mode='lines',
        name='Mean'
    ),
    go.Scatter(
        name='Upper Bound',
        x=dfmax['simtime'],
        y=dfmax['stepdur'],
        mode='lines',
        marker=dict(color="#444"),
        line=dict(width=0),
        showlegend=False
    ),
    go.Scatter(
        name='Lower Bound',
        x=dfmin['simtime'],
        y=dfmin['stepdur'],
        marker=dict(color="#444"),
        line=dict(width=0),
        mode='lines',
        fillcolor='rgba(68, 68, 68, 0.3)',
        fill='tonexty',
        showlegend=False
    )
])

fig.update_layout(
    yaxis_title='Step duration (us)',
    xaxis_title='Step time (s)',
    title=f'Step duration for {obs} tests - with upper/lower bound bands',
    hovermode="x"
)

fig.show()

fig.write_html("log/file.html")
```
